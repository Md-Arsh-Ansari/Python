Display Multiple Plots
With the subplot() function

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])


plt.subplot(1, 2, 1)
plt.plot(x,y, 'o-')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y, 'o-')

plt.show()
```

The subplot() function takes three arguments that describes the layout of the figure.

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.

Draw 2 plots on top of each other:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y, 'o-')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y, 'o-')

plt.show()
```

Title by : title("") and Super Title by: suptitle("")

Size:
You can change the size of the dots with the s argument.
Alpha:
You can adjust the transparency of the dots with the alpha argument.
ColorMap:
    You can specify the colormap with the keyword argument cmap with the value of the colormap
    You can include the colormap in the drawing by including the plt.colorbar() statement:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c = "hotpink", s=sizes, alpha=0.5, cmap= "nipy_spectral")

plt.colorbar()
plt.show()
```

# Bars

Creating Bars
With Pyplot, you can use the bar() function to draw bar graphs:
    
    

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

Horizontal Bars
If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

# Bar color

The bar() and barh() takes the keyword argument color to set the color of the bars:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "SpringGreen")
plt.show()
```

Bar Width:
The bar() takes the keyword argument width to set the width of the bars.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "SpringGreen", width = 0.1)
plt.show()
```

Note: The default width value is 0.8

Note: For "horizontal bars", use height instead of width.

change the Bar Height:
The barh() takes the keyword argument height to set the height of the bars:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color = "LimeGreen", height = 0.1)
plt.show()
```

## Creating Pie Charts:


With Pyplot, you can use the pie() function to draw pie charts:

```python
x = np.array([40, 30, 20, 10])

plt.pie(x)
plt.show() 
```

Note: By default the plotting of the first wedge starts from the x-axis and move counterclockwise:

Labels:
Add labels to the pie chart with the label parameter.

The label parameter must be an array with one label for each wedge.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

Explode:
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

Shadow
Add a shadow to the pie chart by setting the shadows parameter to True:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

Colors:
You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["SpringGreen", "RoyalBlue", "LimeGreen", "Dodgerblue"]
myexplode = [0.2, 0, 0, 0]


plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode)
plt.show() 
```

With Legend:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["SpringGreen", "RoyalBlue", "LimeGreen", "Dodgerblue"]
myexplode = [0.2, 0, 0, 0]


plt.pie(y, labels = mylabels, colors = mycolors, explode = myexplode)
plt.legend(title = "Four Fruits:")
plt.show()
```
